```python
import os
from os import path
import sys
import json
import bz2
from functools import partial
import pickle
import re
import html
```

```python
with open('index.pkl', 'rb') as f:
    page_ids = pickle.load(f)

pages = list(page_ids.keys())
```

```python
source = 'downloads/enwiki-20200401-pages-articles-multistream1.xml-p1p30303.bz2'

count = 0
is_text = False
redirect_page = False
title = None
title_id = None
links = None
ignore_text = False

pattern = re.compile(r"\[\[(.+?)\]\]")

with bz2.BZ2File(source) as f:
    for l in f:
        print(l.decode('utf8'))
        l = l.decode('utf8').strip()
        if l.startswith('<text'):
            is_text = True
            ignore_text = False
            continue
        if l.endswith('</text>'):
            is_text = False
        if l.startswith('<title'):
            title = l[7:-8]
        if l.startswith('<redirect'):
            print(l)
            redirect_page = True
            links.add(l[17:-4].lower())
        if l.startswith('<page'):
            redirect_page = False
            title = ''
            links = set()
        if is_text:
            if redirect_page: print(l)
            l = html.unescape(l.lower())
            print(l)
            if not ignore_text:
                print(l)
                l = re.sub(r"<ref>.*</ref>","<<REF>>",l)
                l = l.replace("_"," ")
                m = pattern.findall(l)
                m = list(map(lambda s : s.split('|')[0].split('#')[0],m))
                links.update(filter(lambda s : s in page_ids,m))
        
        if not redirect_page and l.endswith('</page>'):
            print(title, page_ids[title.lower()])
            print(links)
            print(len(links), 'links')
            print()
            break
```
